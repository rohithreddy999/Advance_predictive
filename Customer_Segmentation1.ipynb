```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Cust_Seg_Train.csv")
df
```

```python
df.drop(columns="ID",axis=1,inplace=True)
```

```python
df=df.rename(columns={"Var_1":"Category"})
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender=le.fit_transform(df.Gender)
df.Ever_Married=le.fit_transform(df.Ever_Married)
df.Graduated=le.fit_transform(df.Graduated)
df.Profession=le.fit_transform(df.Profession)
df.Spending_Score=le.fit_transform(df.Spending_Score)
df.Category=le.fit_transform(df.Category)
df.Segmentation=le.fit_transform(df.Segmentation)
```

```python
df.head()
```

```python
x=df.drop(columns="Segmentation",axis=1)
y=df.Segmentation
```

```python
from sklearn.naive_bayes import CategoricalNB
cnb = CategoricalNB()
cnb.fit(x,y)
```
