**2. Check if there are any duplicate rows in the data. If present, drop them.

```python
if df.duplicated().sum() > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate rows have been removed.")
else:
    print("No duplicate rows found.")
```

**4. List the descriptive statistics for each numeric column.**

```python
numeric_stats = df.describe()
print("\nDescriptive Statistics for Numeric Columns:")
print(numeric_stats)
```

**5. Plot histogram for each numeric column.**

```python
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, bins=20, kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

**6. Plot count plot for each categorical column.**

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column)
    plt.title(f'Count Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

**7. Plot pairplot for all numeric features.**

```python
sns.pairplot(data=df[numeric_columns])
plt.show()
```

**8. Check if the dataset is balanced or imbalanced.**

```python
balance_counts = df['Approved'].value_counts()
print("\nBalance of the dataset:")
print(balance_counts)
```

**9. Plot Heat Map with Correlation Coefficient for all numeric features**

```python
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**10. Plot Kernel Density Estimate plot for Age, CreditScore and YearsEmployed with
target feature Approved as Hue.**

```python
# Subset the dataset to include only the relevant columns
subset_df = df[['Age', 'CreditScore', 'YearsEmployed', 'Approved']]
# Create KDE plots with 'Approved' as Hue
plt.figure(figsize=(12, 6))
for feature in ['Age', 'CreditScore', 'YearsEmployed']:
    sns.kdeplot(data=subset_df, x=feature, hue='Approved', fill=True, common_norm=False, alpha=0.5, label=feature)
plt.title("Kernel Density Estimate (KDE) Plots")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(title='Approved', loc='best')
plt.show()
```

**11. Generate cross table for Gender and Approved columns**

```python
cross_table = pd.crosstab(df['Gender'], df['Approved'])
print("\nCross Table for 'Gender' and 'Approved' columns:")
print(cross_table)
```

**12. Generate bar plots for Ethnicity, Employed and BankCustomer columns with
target feature Approved as Hue.**

```python
for column in ['Ethnicity', 'Employed', 'BankCustomer']:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, hue='Approved')
    plt.title(f'Bar Plot of {column} with Approved as Hue')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

**13. Generate FacetGrid with Married as column, Employed as row, Age as X axis,
YearsEmployed as Y axis and Approved as Hue.**

```python
g = sns.FacetGrid(df, col='Married', row='Employed', hue='Approved', margin_titles=True)
g.map_dataframe(sns.scatterplot, x='Age', y='YearsEmployed')
g.set_axis_labels('Age', 'YearsEmployed')
g.set_titles(col_template="{col_name} Marriage Status", row_template="{row_name} Employed")
g.add_legend()
plt.show()
```
